```python
data['Ram'] = data['Ram'].astype(str).str.replace('GB', '').astype('int32')

data['Weight'] = data['Weight'].astype(str).str.replace('kg', '').astype('float32')
```

```python
def add_company(inpt):
    if inpt=='Samsung' or inpt=='Mediacom' or inpt=='Razer' or inpt=='Microsoft' or inpt=='Vero' or inpt=='Xiaomi' or inpt=='Chuwi' or input=='Fujitsu' or inpt=='Google' or inpt=='LG' or inpt=='Huawei' or inpt=='Fujitsu' :
        return 'Other'

    else:
        return inpt

data['Company'] = data['Company'].apply(add_company)
```

```python
data['cpu_name'] = data['Cpu'].apply(lambda x:" ".join(x.split()[0:3]))
```

```python
data['cpu_name'].value_counts()
```

```python
def set_processor(name):
    if name=="Intel Core i7" or name == 'Intel Core i5' or name == 'Intel Core i3':
        return name
    else:
        if name.split()[0] == 'AMD':
            return 'AMD'
        else:
            return 'Other'

data['cpu_name'] = data['cpu_name'].apply(set_processor)
```

```python
data['cpu_name'].value_counts()
```

```python
data['Gpu'].value_counts()
```

```python
data['gpu_name'] = data['Gpu'].apply(lambda x:" ".join(x.split()[0:1]))
```

```python
data.shape
```

```python
data = data[data['gpu_name'] != 'ARM']
```

```python
data.head(2)
```

```python
data['OpSys'].value_counts()
```

```python
def set_os(inpt):
    if inpt == 'Windows 10' or inpt == 'Windows 7' or inpt == 'Windows 10 S':
        return 'Windows'
    elif inpt == 'macOS' or inpt == 'Mac OS X':
        return 'Mac'
    elif inpt == 'Linux':
        return inpt
    else:
        return 'Other'
        
data['OpSys'] = data['OpSys'].apply(set_os)
```

```python
data = data.drop(columns=['laptop_ID','Inches', 'Product','ScreenResolution','Cpu','Gpu'])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.25)
```

```python
x_train.shape,x_test.shape
```

```python
def model_acc(model):
    model.fit(x_train,y_train)
    acc = model.score(x_test,y_test)
    print(str(model) + '-->' + str(acc))
```

```python
!pip install scikit-learn
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
model_acc(lr)

from sklearn.linear_model import Lasso
lasso = Lasso()
model_acc(lasso)

from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor()
model_acc(dt)

from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
model_acc(rf)
```

```python
from sklearn.model_selection import GridSearchCV
parameters = {
    'n_estimators': [10, 50,100],  # Number of trees
    'criterion': ['squared_error', 'absolute_error','poisson']
}
grid_obj = GridSearchCV(estimator=rf ,param_grid= parameters )
grid_fit = grid_obj.fit(x_train, y_train)

best_model = grid_fit.best_estimator_
best_model
```

```python
best_model.score(x_test,y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV

# Define a larger hyperparameter search space
param_grid = {
    'n_estimators': [50, 100, 200, 300, 500],  
    'max_depth': [None, 10, 20, 30, 40],  
    'min_samples_split': [2, 5, 10],  
    'min_samples_leaf': [1, 2, 4],  
    'criterion': ['squared_error', 'absolute_error']
}

# Perform Randomized Search
rf = RandomForestRegressor(random_state=42)
random_search = RandomizedSearchCV(estimator=rf, param_distributions=param_grid, 
                                   n_iter=20, cv=5, n_jobs=-1, verbose=2, random_state=42)

random_search.fit(x_train, y_train)

# Best model after tuning
best_rf = random_search.best_estimator_
best_rf
# Test the model accuracy again
#best_rf_score = best_rf.score(x_test, y_test)
#print("Optimized RandomForestRegressor Score:", best_rf_score)
```

```python
best_rf.score(x_test,y_test)
```
